```python
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.ticker import *

plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python

def add_zero(date_string):
    arr = list(date_string)
    if len(date_string) == 6:
        arr = arr + ['0']
        arr[-1] = arr[-2]
        arr[-2] = '0'
    return ''.join(arr)

df_py = pd.read_csv('./QueryResults_python.csv')
df_js = pd.read_csv('./QueryResults_javascript.csv')

df_py['dt'] = df_py['dt'].map(add_zero)
df_js['dt'] = df_js['dt'].map(add_zero)

df_py = df_py.sort_values(by='dt')
df_js = df_js.sort_values(by='dt')
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12,5))
ax = plt.axes()

plt.plot(df_py['dt'], df_py['num_mentions'], label='python')
plt.plot(df_js['dt'], df_js['num_mentions'], label='javascript')
plt.xticks(rotation=60)
plt.rcParams.update({'font.size': 12})

plt.legend()
ax.xaxis.set_major_locator(MultipleLocator(10))
```
